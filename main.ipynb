```python
import pandas as pd

df = pd.read_csv('Preload_simulation_data/CURVE_1_X_Y.csv', skiprows=1)
df.columns = ['Candidate', 'PUL stiffness', 'pul', 'Pressure Magnitude MPa', 
              'Moment X Nm', 'Moment Total Nm', 'P_point_Y_Avg mm', 'P_point_Z_Avg mm']

df['Moment X Nm'] = df['Moment X Nm'] * 1000
df['Moment Total Nm'] = df['Moment Total Nm'] * 1000
df.rename(columns={'Moment X Nm': 'Moment X Nmm',
                   'Moment Total Nm': 'Moment Total Nmm'},
          inplace=True)

```

```python
low_preassure = df[df['Pressure Magnitude MPa'] <= 0.0006]
middle_preassure = df[(df['Pressure Magnitude MPa'] > 0.0006) & (df['Pressure Magnitude MPa'] <= 0.0010)]
high_preassure = df[df['Pressure Magnitude MPa'] > 0.0010]
```

```python
low_p_motion = low_preassure[['PUL stiffness', 'Moment Total Nmm', 'P_point_Y_Avg mm', 'P_point_Z_Avg mm']]
middle_p_motion = middle_preassure[['PUL stiffness', 'Moment Total Nmm', 'P_point_Y_Avg mm', 'P_point_Z_Avg mm']]
high_p_motion = high_preassure[['PUL stiffness','Moment Total Nmm', 'P_point_Y_Avg mm', 'P_point_Z_Avg mm']]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Filtrar low pressure y seleccionar (ajusta columna deformacin)
low_df = low_preassure # Ej. baja presin
deform_p = low_df['P_point_Z_Avg mm']  # O 'Total def Max', tu P point
stiff_pul = low_df['PUL stiffness']

plt.figure(figsize=(9, 6))
plt.scatter(deform_p, stiff_pul, color='red', s=80, alpha=0.7)
plt.plot(deform_p, stiff_pul, 'b--', linewidth=2)  # Lnea conectada
plt.ylabel('PUL Stiffness (N/mm)')
plt.xlabel('Deformacin P Point (mm)')
plt.title('Deformacin P Point vs PUL Stiffness (Low Pressure)')
plt.grid(True, alpha=0.3)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

low_df = low_preassure  # DataFrame de baja presin

# Columnas
p_z = low_df['P_point_Z_Avg mm']
p_y = low_df['P_point_Y_Avg mm']
stiff_pul = low_df['PUL stiffness']

# ngulo (en radianes)
angle = np.arctan((7.5 - p_z) / (22.5 + p_y))

angle = np.degrees(angle)

plt.figure(figsize=(9, 6))
plt.scatter(angle, stiff_pul, color='red', s=80, alpha=0.7)
plt.plot(angle, stiff_pul, 'b--', linewidth=2)

plt.ylabel('PUL Stiffness (N/mm)')
plt.xlabel('Angle (rad)')
plt.title('PUL Stiffness over Angle (Low Pressure)')
plt.grid(True, alpha=0.3)
plt.show()
```
